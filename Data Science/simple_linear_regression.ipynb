# Introduction to Regression 

## What is Regression?
Regression is a statistical method used to find relationships between variables. It is widely used to:
- Understand how one factor influences another.
- Make predictions based on available data.

For example:
- **In business**: How does advertising affect sales?
- **In education**: How does the number of study hours affect exam performance?
- **In healthcare**: How does age influence blood pressure?

---

## Why Learn Regression?
1. **Real-World Applications**: Regression is used in almost every field, such as finance, marketing, healthcare, and sports.
2. **Easy to Understand**: It involves basic concepts like drawing a line or curve to show relationships.
3. **Foundation for Advanced Techniques**: Many advanced machine learning techniques build on regression.

---

## Key Concepts in Regression
### Dependent Variable
- This is the variable we want to predict or explain.
- Example: Test scores in exams.

### Independent Variable
- This is the variable that influences the dependent variable.
- Example: Hours spent studying.

### Regression Line
- A straight line that best fits the data points.
- Shows the trend or relationship between the variables.

### Equation of a Regression Line
The equation of a straight line is:
Y = mX + b



- **Y**: Predicted value (e.g., test score)
- **X**: Independent variable (e.g., hours of study)
- **m**: Slope of the line (how steep the line is)
- **b**: Intercept (where the line crosses the Y-axis)

---

## Types of Regression
### 1. Linear Regression
- Used when the relationship between variables is linear (a straight line).
- **Example**: Predicting test scores based on study hours.

### 2. Multiple Linear Regression
- Used when there are multiple independent variables.
- **Example**: Predicting house price based on size, location, and number of bedrooms.

### 3. Logistic Regression
- Used for predicting categories (yes/no, pass/fail).
- **Example**: Will a student pass the exam? (Yes or No)

---

## How Does Linear Regression Work?
Linear regression tries to find the "best-fit line" by minimizing the error. This error is the difference between the actual and predicted values.

### Example Dataset
| Hours of Study | Test Score |
|----------------|------------|
| 1              | 50         |
| 2              | 55         |
| 3              | 60         |
| 4              | 65         |
| 5              | 70         |

- **Question**: If a student studies for 6 hours, what will their test score be?
- **Answer**: Use the regression line to predict the score.

---

## Visualizing Regression
A graph can help visualize the relationship between variables. For example:
- X-axis: Hours of study
- Y-axis: Test score
- The regression line shows the trend.

# Simple Linear Regression

## Importing the libraries

````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
````

````python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Step 1: Prepare Data
data = {
    "Hours_of_Study": [1, 2, 3, 4, 5],
    "Test_Score": [50, 55, 60, 65, 70]
}
df = pd.DataFrame(data)

# Step 2: Separate Independent and Dependent Variables
X = df[["Hours_of_Study"]]
Y = df["Test_Score"]

# Step 3: Train-Test Split
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)

# Step 4: Create Linear Regression Model
model = LinearRegression()
model.fit(X_train, Y_train)

# Step 5: Make Predictions
test_predictions = model.predict(X_test)

# Step 6: Visualize the Results
# Scatter plot of actual data
plt.scatter(X, Y, color="blue", label="Actual Data")
plt.plot(X, model.predict(X), color="red",
         label="Regression Line")  # Regression line

# Mark the test predictions on the plot
for i, (x, y_true, y_pred) in enumerate(zip(X_test.values, Y_test, test_predictions)):
    plt.scatter(x, y_pred, color="orange",
                label="Test Prediction" if i == 0 else "")
    plt.text(x, y_pred, f"{y_pred:.2f}", color="orange", fontsize=9)

# Step 7: Predict an Unknown Value
new_data = [[6]]  # New input for prediction
predicted_score = model.predict(new_data)

# Add the new prediction to the plot
plt.scatter(new_data, predicted_score, color="green", label="New Prediction")
plt.text(new_data[0][0], predicted_score[0], f"{
         predicted_score[0]:.2f}", color="green", fontsize=12)

# Customize plot
plt.xlabel("Hours of Study")
plt.ylabel("Test Score")
plt.title("Linear Regression with Predictions")
plt.legend()
plt.show()

# Print predictions for clarity
print(f"\nPredicted Test Scores for Test Data: {test_predictions}")
print(f"Predicted Test Score for 6 Hours of Study: {predicted_score[0]}")
````

## Importing the dataset

````python
dataset = pd.read_csv('Salary_Data.csv')
print(dataset.head)
X = dataset.iloc[:, :-1].values # all rows apart from last column
y = dataset.iloc[:, -1].values # all rows of last column
````

## Splitting the dataset into the Training set and Test set

````python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
````

## Training the Simple Linear Regression model on the Training set

````python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
````

## Predicting the Test set results

````python
y_pred = regressor.predict(X_test)
````

## Visualising the Training set results

````python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
````

## Visualising the Test set results

````python
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
````

````python
years_of_experience = np.array([[6]])  # Reshape it to 2D as the model expects a 2D array
predicted_salary = regressor.predict(years_of_experience)

print(f"Predicted salary for 6 years of experience: {predicted_salary[0]}")
````
