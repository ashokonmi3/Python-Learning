### Pivoting DataFrames with pivot()

What is a Pivot Table?

A pivot table is a data summarization tool used in data analysis, often found in software like Excel, pandas (Python), or Google Sheets. It allows you to organize and summarize large datasets by grouping data in a meaningful way, and calculating aggregates (like sum, mean, count, etc.) for specific subsets of the data.

In a pivot table, you can:

Group data by one or more columns (e.g., by category, date, or location).
Calculate statistics or aggregations for each group.
Rearrange the data to get different views, helping you to understand patterns and insights more clearly.
Real-Life Analogy:
Imagine you're a teacher and you want to know how many students passed or failed by subject. You have a list of students' scores, but it’s hard to figure out the total number of students passing or failing by subject without sorting and summarizing the data manually.

With a pivot table, you can easily:

Group students by subject.
Calculate the pass/fail rate for each subject.
Summarize all the data in one table.
How Does a Pivot Table Work in Pandas (Python)?
In pandas, a pivot table works similarly to Excel, where you specify:

Rows: The column you want to group data by.
Columns: The column for which you want to calculate aggregates.
Values: The data you want to calculate (such as sum, mean, etc.).

### Limits of pivot()

### Handling Multiple Aggregations
In pandas, pivot tables are designed to work with a single aggregation function at a time (e.g., sum, mean). If you need to perform multiple aggregations (like calculating both the sum and mean for the same column), the pivot table can’t directly support this without additional work.

```python
import pandas as pd

# Sample DataFrame
data = {'Country': ['USA', 'USA', 'USA', 'Canada', 'Canada', 'Mexico'],
        'Medal': ['Gold', 'Silver', 'Bronze', 'Gold', 'Silver', 'Gold'],
        'Count': [10, 5, 3, 4, 2, 7]}

df = pd.DataFrame(data)

# Pivot Table: Trying to calculate sum and mean in one table
pivot = df.pivot_table(values='Count', index='Country',
                       columns='Medal', aggfunc='sum')

# Display the pivot table
print(pivot)
```

This pivot table only allows one aggregation function, such as sum. If you wanted to calculate both the sum and mean for each combination of Country and Medal, you would need to either:

Create two separate pivot tables (one for sum, one for mean).

Use a more complex aggregation method.

### Handling Missing Data
If your dataset contains missing values (e.g., certain countries don’t have medals in a certain category), a pivot table will display NaN (Not a Number) by default for those missing values. Although you can use the fill_value parameter to replace NaN with a value, it is not as flexible or intuitive as handling missing data directly in other ways (such as using groupby).

```python
# Sample DataFrame with missing data (NaN)
data = {'Country': ['USA', 'USA', 'USA', 'Canada', 'Mexico'],
        'Medal': ['Gold', 'Silver', 'Bronze', 'Gold', 'Gold'],
        'Count': [10, 5, 3, 4, 7]}

df = pd.DataFrame(data)

# Pivot Table: Using fill_value to replace NaN with 0
pivot = df.pivot_table(values='Count', index='Country',
                       columns='Medal', aggfunc='sum', fill_value=0)

# Display the pivot table
print(pivot)
```

3. Cannot Handle Complex Aggregations Without Custom Functions
Pivot tables are great for simple aggregations (like sum, mean, count), but they can’t handle more complex custom calculations unless you use custom aggregation functions.

Limitation Example: Custom Aggregations:
Suppose we want to calculate the percentage of total medals each country won, per medal type (e.g., what percentage of the total gold medals did Canada win?).

To do this with a pivot table, you'd need to write a custom aggregation function, but it's not as straightforward as applying a basic aggregation function like mean or sum.

```python
# Sample DataFrame with counts of medals
data = {'Country': ['USA', 'USA', 'USA', 'Canada', 'Canada', 'Mexico'],
        'Medal': ['Gold', 'Silver', 'Bronze', 'Gold', 'Silver', 'Gold'],
        'Count': [10, 5, 3, 4, 2, 7]}

df = pd.DataFrame(data)

# Pivot Table: Trying to calculate percentages (not directly possible)
pivot = df.pivot_table(values='Count', index='Country',
                       columns='Medal', aggfunc='sum', fill_value=0)

# Display the pivot table
print(pivot)

# Now, let's calculate the percentage of gold medals won by each country manually
total_medals = pivot.sum(axis=1)
pivot_percentages = pivot.divide(total_medals, axis=0) * 100

print("\nPercentage of Medals:")
print(pivot_percentages)
```

### pd.crosstab()

pd.crosstab() is a function in pandas used to create contingency tables (also known as cross-tabulations). It is a quick way to summarize the relationship between two or more categorical variables. It shows the frequency or count of different combinations of categories in the dataset.

In simple terms, pd.crosstab() helps you compare the occurrence of categories from one or more columns and represents it in a table form, making it easier to analyze the relationships between those columns.

Syntax:

pd.crosstab(index, columns, values=None, aggfunc=None, normalize=False)

index: The column (or series) to be used as the rows of the resulting table.

columns: The column (or series) to be used as the columns of the resulting table.

values (optional): The column to aggregate if you want to calculate a summary statistic (like sum, mean, etc.) instead of just counting occurrences.

aggfunc (optional): The aggregation function to apply (e.g., sum, mean).
normalize (optional): If set to True, it normalizes the values to show proportions instead of counts.
