### Ranking DataFrames with rank()

````python
import pandas as pd
````

````python
sales = pd.Series([15, 32, 45, 21, 55, 15, 0],  index = ["Mo", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"])
````

````python
sales = pd.Series([15, 32, 45, 15, 55, 15, 0],  index = ["Mo", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"])
````

````python
sales
````

````python
sales.sort_values(ascending = False)
````

````python
sales.rank(ascending=False, method = "min").sort_values(ascending = True)
# ascending=False: Ranks in descending order, i.e., higher values get lower rank numbers.
# method="min": Assigns the lowest rank to tied values.( 2 equal rank)
# 5,6 rank wont be given
````

````python
sales.rank(ascending=False, method = "min", pct=True).sort_values()
# ascending=False: Ranks in descending order (i.e., higher values get lower ranks).
# method="min": For tied values, assigns the smallest rank among the tied group.
# pct=True: Converts ranks into percentile ranks (relative position of the value in the dataset).
# Percentile rank ranges between 0 and 1.
# Example: A rank of 1 in a list of 5 items corresponds to 1/5 = 0.2.
````

````python
titanic = pd.read_csv("titanic.csv")
````

````python
titanic.fare.rank(ascending = False)
````

````python
titanic["fare_rank"] = titanic.fare.rank(ascending = False, method="min")
````

````python
titanic.head()
````

````python
titanic.sort_values("fare", ascending= False)
````

````python
titanic.drop(columns = "fare_rank", inplace= True)
````

# Differences Between `sort` and `rank`

| **Feature**            | **Sort**                                 | **Rank**                                |
|-------------------------|------------------------------------------|-----------------------------------------|
| **Purpose**             | Rearranges the data in order.            | Assigns a numerical rank to each value. |
| **Output**              | Sorted data.                             | Original data with a rank column/series.|
| **Data Order**          | Changes the order of the data.           | Preserves the original order.           |
| **Ties (Same Value)**   | Doesn't affect the order of tied values. | Ties are handled with ranking methods (`min`, `max`, `average`, etc.). |

# `apply()`, `map()`, and `applymap()` in Pandas

In data analysis, we often need to perform operations on data in a DataFrame or Series. Pandas provides three powerful functions for this purpose: `apply()`, `map()`, and `applymap()`. Here's a simple explanation of each for non-technical people:

---

## **1. `map()`**
The `map()` function is used for applying a function or operation to **each element in a Pandas Series**.

### **When to Use:**
- Use `map()` when you want to modify or transform values in a **single column** (or Series).

### **Example Use Cases:**
- Add a fixed value to every number in a column.
- Replace specific values (e.g., change "Male" to 0 and "Female" to 1).

---

## **2. `apply()`**
The `apply()` function is used to apply a function along **rows or columns** of a DataFrame or to each element in a Series.

### **When to Use:**
- Use `apply()` when you need to:
  - Perform row-wise or column-wise operations in a DataFrame.
  - Apply complex functions to a Series or DataFrame.

### **Example Use Cases:**
- Calculate the sum of all values in each row.
- Convert all text in a column to lowercase.

---

## **3. `applymap()`**
The `applymap()` function is used to apply a function **element-wise to all cells** in a DataFrame.

### **When to Use:**
- Use `applymap()` when you want to modify **every single value in a DataFrame**.

### **Example Use Cases:**
- Multiply all numbers in a DataFrame by 2.
- Round all numerical values to two decimal places.

---

## **Key Differences**

| **Function**   | **Applies To**           | **Scope**                     | **When to Use**                                   |
|-----------------|--------------------------|--------------------------------|--------------------------------------------------|
| `map()`         | **Pandas Series**       | Each element in the Series    | Modify or transform a single column/Series.     |
| `apply()`       | **Series or DataFrame** | Rows, columns, or elements    | Perform operations row-wise or column-wise.     |
| `applymap()`    | **DataFrame**           | Every element in the DataFrame| Modify all values in a DataFrame.               |

---
