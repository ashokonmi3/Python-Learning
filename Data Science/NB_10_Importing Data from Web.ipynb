```python
import pandas as pd

# URL of the page containing the table
url = "https://en.wikipedia.org/wiki/1976_Summer_Olympics_medal_table"

# Read all tables from the URL
tables = pd.read_html(url)

# Check how many tables are on the page
print(f"Number of tables found: {len(tables)}")

# Access the 3rd table (index 3)
third_table = tables[3]

# Display the 3rd table
print(third_table)
```
