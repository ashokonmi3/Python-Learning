```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 20, 25]

# Plot
plt.plot(x, y)

# Labels and Title
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Simple Line Plot")

# Show the plot
plt.show()
```

4. Adding Markers, Line Styles, and Colors


```python
plt.plot(x, y, marker='o', linestyle='--',
         color='r', markersize=8, linewidth=2)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Customized Line Plot")
plt.show()
```

- `marker='o'` → Adds circle markers.
- `linestyle='--'` → Dashed line.
- `color='r'` → Red line.
- `markersize=8` → Size of markers.
- `linewidth=2` → Line thickness.


Creating a Bar Chart
- `plt.bar()` → Creates a bar chart.
- `color=[]` → Specifies bar colors.


```python
categories = ["A", "B", "C", "D"]
values = [5, 7, 3, 8]

plt.bar(categories, values, color=['blue', 'green', 'red', 'purple'])
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart Example")
plt.show()
```

### Creating a Scatter Plot
- `plt.scatter()` → Creates a scatter plot.
- `marker='*'` → Star markers.
- `s=100` → Size of markers.


```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.scatter(x, y, color='blue', marker='*', s=100)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Scatter Plot Example")
plt.show()
```

### Creating a Histogram
- `plt.hist()` → Creates a histogram.
- `bins=30` → Number of bins (intervals).
- `edgecolor='black'` → Adds black borders to bars.


```python
import numpy as np

data = np.random.randn(1000)  # Generate 1000 random numbers

plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram Example")
plt.show()
```

### Creating a Pie Chart
- `autopct='%1.1f%%'` → Shows percentage values.
- `startangle=140` → Rotates the chart.
startangle parameter is used to rotate the chart. It determines the angle at which the first segment (slice) of the pie chart starts. The value is specified in degrees.

```python
labels = ["Apple", "Banana", "Cherry", "Date"]
sizes = [30, 20, 40, 10]
colors = ["red", "yellow", "pink", "brown"]

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart Example")
plt.show()
```

## Adding Multiple Lines in a Single Plot
- `plt.legend()` → Displays labels.


```python
x = [1, 2, 3, 4, 5]
y1 = [10, 15, 20, 25, 30]
y2 = [5, 10, 15, 20, 25]

plt.plot(x, y1, label="Line 1", color='blue', marker='o')
plt.plot(x, y2, label="Line 2", color='red', linestyle='--', marker='s')

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Multiple Lines in One Plot")
plt.legend()  # Shows legend
plt.show()
```

### Customizing Axes and Grid
- `plt.xlim()` → Sets the limits for the x-axis. You can specify the minimum and maximum values for the x-axis. 
  - Example: `plt.xlim(0, 10)` will set the x-axis to range from 0 to 10.
  
- `plt.ylim()` → Sets the limits for the y-axis. You can specify the minimum and maximum values for the y-axis.
  - Example: `plt.ylim(-5, 5)` will set the y-axis to range from -5 to 5.

- `plt.grid()` → Adds a grid to the plot. The grid makes it easier to read values from the plot. By default, it adds a grid with both x and y axis lines.
  - Example: `plt.grid(True)` enables the grid, and `plt.grid(False)` disables it.
  - You can also customize the grid style:
    - `plt.grid(color='gray', linestyle='--', linewidth=0.5)` → Adds a gray, dashed grid with a thinner line.

You can combine `plt.xlim()` and `plt.ylim()` to control both axes' ranges while enabling a grid for better visualization.


```python
plt.plot(x, y1, color='purple', linestyle='-.', marker='^')

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Plot with Grid and Limits")

plt.xlim(0, 6)  # Set X-axis range
plt.ylim(5, 35)  # Set Y-axis range
plt.grid(True, linestyle='--', alpha=0.7)  # Add grid

plt.show()
```

### Saving a Plot as an Image
- `plt.savefig("filename.png", dpi=300)` → Saves the plot as an image file.
  - `"filename.png"` → Specifies the name and format of the image file (you can choose `.png`, `.jpg`, `.svg`, `.pdf`, etc.).
  - `dpi=300` → Sets the resolution of the saved image. Higher DPI values (like 300) provide higher resolution and better quality, especially for printing. For web use, a DPI of 72 to 150 is often sufficient.


```python
plt.plot(x, y1, color='blue')
plt.title("Saving a Plot")

plt.savefig("plot.png", dpi=300)  # Save plot as an image
plt.show()

```
