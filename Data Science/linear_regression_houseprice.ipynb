```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Creating the dataset
data = {'Area': [2600, 3000, 3200, 3600, 4000],
        'Price': [550000, 565000, 610000, 680000, 725000]}
df = pd.DataFrame(data)

# Scatter plot of the actual data points
plt.figure(figsize=(10, 6))
plt.scatter(df.Area, df.Price, color='red', marker='+', label='Actual Data')

# Setting up the Linear Regression model
reg = LinearRegression()
reg.fit(df[['Area']], df.Price)

# Predicting prices for the given areas to plot the regression line
predicted_price = reg.predict(df[['Area']])
plt.plot(df.Area, predicted_price, color='blue', label='Regression Line')

# Predict price for 3300 sq ft using the trained model
predicted_value = reg.predict([[3300]])

# Plot the predicted point for 3300 sq ft (green dot)
plt.scatter(3300, predicted_value, color='green', s=100,
            label=f"Predicted Price for 3300 sq ft: {predicted_value[0]:.2f} US$")

# Customizing the plot
plt.xlabel('Area (sqr ft)')
plt.ylabel('Price (US$)')
plt.title('Linear Regression: Area vs Price')
plt.legend()
plt.grid()
plt.show()

# Print the predicted price for 3300 sq ft
print(f"Predicted Price for 3300 sq ft: {predicted_value[0]:.2f} US$")
```
